```python
import pandas as pd
from zipfile import ZipFile
import seaborn as sns 
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Opening the zip file and extracting the dataset
ZipFile("pd_nibrs_datasd.zip").extractall()

# Read the dataset into a DataFrame
df = pd.read_csv('pd_nibrs_datasd.csv', low_memory=False)

# Create a new DataFrame keeping only the relevant data
data = df[['case_number', 'occured_on', 'code_section', 'pd_offense_category',
                'neighborhood', 'division', 'block_addr', 'city', 'zip', 'latitude', 'longitude']]
```

```python
data.head()
```

```python
# Display basic info about the DataFrame
print("DataFrame Info:")
data.info()

# Checking for missing values in each column
print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python
# Convert 'occurred_on' to datetime format and extract date components in one go
data['occurred_on'] = pd.to_datetime(data['occured_on'], errors='coerce')

# Extract year, month, day of week, day, hour, and minute
data[['year', 'month', 'day_of_week', 'day', 'hour', 'minute']] = data['occurred_on'].apply(
    lambda x: pd.Series([x.year, x.month, x.dayofweek, x.day, x.hour, x.minute] if pd.notnull(x) else [None]*6)
)

# Drop the original 'occured_on' column
data = data.drop(columns=['occured_on'])
```

```python
# Fill NaN values with a placeholder, for example, '0'
data['zip'] = data['zip'].fillna(0).astype(int).astype(str)
```

```python
# Drop rows that still have missing values in 'latitude' or 'longitude'
data = data.dropna(subset=['latitude', 'longitude'])
```

```python
# Remove trailing '||' from code_section
data['code_section'] = data['code_section'].str.replace(r' \|\| $', '', regex=True)
```

```python
# Make block_addr title case
data['block_addr'] = data['block_addr'].str.title()
```

```python
# Display basic info about the DataFrame
print("DataFrame Info:")
data.info()

# Checking for missing values in each column
print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python
data.head()
```

```python
# Export the cleaned data to a new CSV
data.to_csv('cleaned_crime_data.csv', index=False)

# Check the result
print("Data has been cleaned and saved as 'cleaned_crime_data.csv'.")
```
